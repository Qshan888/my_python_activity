```python
movies = pd.read_excel("movies.xlsx")

print(movies.shape)
print(movies.head())
print(movies.tail())
print(movies.dtypes)

```

```python
movies['Age'].isnull().sum()
```

```python
movies['Age']
```

```python
# Replace the missing values in Age column with a set value
movies['Age'].fillna("Others", inplace= True)
```

```python
movies['Directors'].isnull().sum()
```

```python
movies['Directors'].isna().sum()
```

```python
# Replace the missing values in Age column with a set value
movies['Directors'].fillna("Others", inplace= True)
movies['Genres'].fillna("Others", inplace= True)
movies['Country'].fillna("Others", inplace= True)
movies['Language'].fillna("Others", inplace= True)

movies
```

```python
# Missing values for IMDb and Rotten Tomatoes
print(movies['IMDb'].isnull().sum())

print(movies['Rotten Tomatoes'].isnull().sum())
```

```python
# Replace missing values with mean
movies['IMDb'].fillna(movies['IMDb'].mean(),inplace=True)

movies['Rotten Tomatoes'].fillna(movies['Rotten Tomatoes'].mean(),inplace=True)
```

```python
# View the missing values for IMDb and Rotten Tomatoes
print(movies['IMDb'].isnull().sum())

print(movies['Rotten Tomatoes'].isnull().sum())
```

```python
movies.dtypes
```

```python
# Filter the DataFrame according to string variables.
movies_string = movies.select_dtypes('object')
movies_string
```

```python
# Filter the DataFrame according to numeric variables.
movies_num = movies.select_dtypes('number')
movies_num = movies_num.drop('Year', axis = 1)
```

```python
movies_num
```

```python
Q1 = movies_num.quantile(0.25)
Q3 = movies_num.quantile(0.75)

print(Q1)
print(Q3)
```

```python
# Calculate the IQR.
IQR = Q3 - Q1

# View the output.
print(IQR)
```
